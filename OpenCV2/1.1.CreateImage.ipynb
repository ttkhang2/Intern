```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.zeros([60, 60, 3])
h, w, c = data.shape

color_map = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (0, 0, 0), (255, 255, 255),
             (100, 100, 100), (120, 50, 10), (36, 167, 227), (240, 212, 0),]
```

```python
assert h == w
step = h//3
i=0
for row in range(3):
    for col in range(3):
        data[row*step:(row+1)*step, col*step:(col+1)*step,:] = color_map[i]
        i+=1

cv2.imwrite("Image/simple.png",data)

plt.imshow(data)
```
